```python
# Import Dependencies 
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style(style="whitegrid")
```

```python
# Import data set
df=pd.read_csv("cardio_df_ultimate.csv")
```

```python
# Remove second index column
df = df.drop(columns=["Unnamed: 0"], axis=1)
df.head(3)
```

```python
# checking bins for systolic_bp to make graphs easier to interpret

df.systolic_bp.value_counts(bins=11)
```

```python
# making a new DF for systolic bins (12 of them)

df_bins = df
```

```python
# df_bins["systolic_bins"] = pd.cut(df_bins["systolic_bp"].values, bins=11, labels=False)
# df_bins.head(3)
```

```python
df_1= df_bins.sample(n=350)
df_1.head(5)
```

```python
# Scatterplot for weight and systolic_bp
# sns.lmplot(
#     data=penguins,
#     x="bill_length_mm", y="bill_depth_mm", hue="species",
#     height=5)
sns.lmplot(data=df_1,
            x="BMI", y="cardio", hue="gender", palette="coolwarm",
          height=7)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(9.5,9.5))
sns.histplot(data=df_1, x="BMI", hue="cardio", kde=True, bins=15, palette="Accent")
plt.title('Effect of high BMI on "Cardio" feature', fontsize=20)
plt.ylabel("Count", fontsize=18)
plt.xlabel("Body Mass Index", fontsize=18)

plt.show()
```

```python
# Relationship between BMI x Cholesterol Level | Polygon | 

f, ax = plt.subplots(figsize=(9.5,9.5))
sns.histplot(data=df_1, x= "BMI", hue="cholesterol", palette="brg", element="poly")
plt.title("Relationship between BMI and Cholesterol Level", fontsize=20)
plt.ylabel("Count", fontsize=18)
plt.xlabel("BMI (Body Mass Index) %", fontsize=18)

# Shows the positive correlation between BMI and Cholesterol level.  
```

```python
# Correlational Heat Map |

correlation = df_1.corr()
f, ax = plt.subplots(figsize = (13,11.5))
sns.heatmap(correlation, annot=True, fmt=".3f", linewidths=0.5, ax=ax, 
            cmap="RdYlGn_r")
plt.title("Correlation Heat Map", fontsize=18)
```

```python
# Diastolic BP x BMI Bar Chart


# # bins = ["89.91-98.0", "98.1-106", "106.1-114.2", "114.3-122.3", "122.4-130.4",
# #        "130.5-138.5", "138.6-146.6", "146.7-154.7", "154.8-162.8", "162.9-170.8",
# #        "170.9-179"]
sns.catplot(data=df_1, x= "diastolic_bp", y="BMI", hue="gender", palette="viridis", kind="bar", errwidth=.05)
plt.title("Effect on Diastolic BP based on BMI", fontsize=20)
plt.ylabel("BMI (Body Mass Index)", fontsize=12)
plt.xlabel("Diastolic Blood Pressure", fontsize=12)

plt.show()
```

```python
# Effect on Systolic BP based on BMI | Bar Graph | 

f, ax = plt.subplots(figsize=(9.5, 9.5))

sns.barplot(data=df_1, x= "systolic_bp", y="BMI", palette="mako")
plt.title("Effect on Systolic BP based upon BMI", fontsize=20)
plt.ylabel("BMI (Body Mass Index)", fontsize=18)
plt.xlabel("Systolic Blood Pressure", fontsize=18)
# x_ticks = np.arange(80,190,15)
# plt.xticks(x_ticks)

plt.show()
```

```python
sns.catplot(data=df_1, x= "systolic_bp", y="BMI", hue="gender", palette="mako", kind="bar", errwidth=.05)
plt.title("Effect on Systolic BP based upon BMI", fontsize=20)
plt.ylabel("BMI (Body Mass Index)", fontsize=12)
plt.xlabel("Systolic Blood Pressure", fontsize=12)
plt.figure(figsize=(20,10))
# x_ticks = np.arange(80,190,15)
# plt.xticks(x_ticks)

plt.show()
```
